```python
import numpy as np
import matplotlib.pyplot as plt

path = '../data/RAR files/'

f = open(path + 'D61DBLREMOPT.txt', 'r')
for i in range(18):
	f.readline()

eeg_list = []
timestamps_list = []
for line in f:
	line = line.rstrip()
	elems = line.split('\t')
	timestamps_list.append(elems[0].split(' ')[2].split(':')[2])
	eeg_list.append(elems[1])
    
eeg = np.array(eeg_list)

wsize = 1280   # window size. data is sampled at 128 Hz, so 1280 sample points = 10 sec.

stime = 0
spect_list = []
power_ar = np.empty((0, wsize), float)
while stime < eeg.shape[0]:
	etime = stime + wsize
	ps = [np.abs(np.fft.fft(eeg[stime:etime])) ** 2]
	power_ar = np.append(power_ar, ps, axis = 0)
	stime = etime

time_step = 1 / 30
freqs = np.fft.fftfreq(power_ar.shape[1], time_step)
idx = np.argsort(freqs)

plt.plot(freqs[idx], ps[idx])
```

```python
import numpy as np
import matplotlib.pyplot as plt

origFreq = 25
wsizeInSec = 10
samplingFreq = 128

sinc = np.sin(np.linspace(0, origFreq * 2 * np.pi * wsizeInSec, samplingFreq * wsizeInSec))
timeInSec = np.linspace(0, wsizeInSec, samplingFreq * wsizeInSec)
plt.plot(timeInSec, sinc)

wsize = samplingFreq * wsizeInSec   # window size. data is sampled at 128 Hz, so 1280 sample points = 10 sec.

ps = np.abs(np.fft.fft(sinc)) ** 2

time_step = 1 / samplingFreq
freqs = np.fft.fftfreq(ps.size, d = time_step)
idx = np.argsort(freqs)

plt.plot(freqs[idx], ps[idx])

```
